## Dataset Statistics

```python
DATASETS = ["Children", "Computers", "Fitness", "History", "Photo", "amazon-ratings", "minesweeper", "questions", "roman-empire", "tolokers"]
DATASETS_SHOW_NAME = ["Children", "Comp.", "Fitness", "History", "Photo", "Amazon.", "Mines.", "Questions", "Roman.", "tolokers"]

RATIO = [0.5, 0.8, 0.7, 0.5, 0.6, 0.6, 0.8, 0.9, 0.2, 0.7]
DATASET_PATH = "../data"
DATASET_HOMOPHILY_PATH = "../data/TFI"
MASK = 0
```

```python
import numpy as np
import pandas as pd
df_data = []
for i, dataset in enumerate(DATASETS):
    dataset_data = np.load(f"{DATASET_PATH}/{dataset.replace('-','_')}.npz")
    # print(f"{dataset} {dataset_data.keys()}")
    dataset_feature = dataset_data['node_features']
    dataset_all_feature_mean = np.mean(dataset_feature)
    dataset_all_feature_std = np.std(dataset_feature)
    dataset_all_feature_sparseness = np.mean(np.sum(dataset_feature == 0, axis=0) / dataset_feature.shape[0])
    tfi_data = np.load(f"{DATASET_HOMOPHILY_PATH}/{dataset.replace('-','_')}_mi_agg_z-norm_mask_{MASK}.npy")
    dataset_all_TFI_mean = np.mean(tfi_data)
    dataset_all_TFI_std = np.std(tfi_data)
    # sort by TFI
    # print(tfi_data.shape)
    sorted_indices = np.argsort(tfi_data)[::-1]
    sorted_node_features_all = dataset_feature[:, sorted_indices.copy()]
    ratio = RATIO[i]
    dataset_G_feature = sorted_node_features_all[:int(ratio * dataset_feature.shape[0])]
    dataset_G_feature_mean = np.mean(dataset_G_feature)
    dataset_G_feature_std = np.std(dataset_G_feature)
    dataset_G_feature_sparseness = np.mean(np.sum(dataset_G_feature == 0, axis=0) / dataset_G_feature.shape[0])
    dataset_sorted_tfi = tfi_data[sorted_indices.copy()]
    dataset_G_TFI_mean = np.mean(dataset_sorted_tfi[:int(ratio * dataset_feature.shape[1])])
    dataset_G_TFI_std = np.std(dataset_sorted_tfi[:int(ratio * dataset_feature.shape[1])])

    dataset_NG_feature = sorted_node_features_all[int(ratio * dataset_feature.shape[0]):]
    dataset_NG_feature_mean = np.mean(dataset_NG_feature)
    dataset_NG_feature_std = np.std(dataset_NG_feature)
    dataset_NG_feature_sparseness = np.mean(np.sum(dataset_NG_feature == 0, axis=0) / dataset_NG_feature.shape[0])
    dataset_NG_TFI_mean = np.mean(dataset_sorted_tfi[int(ratio * dataset_feature.shape[1]):])
    dataset_NG_TFI_std = np.std(dataset_sorted_tfi[int(ratio * dataset_feature.shape[1]):])
    # break
    df_data.append([DATASETS_SHOW_NAME[i], f"{dataset_all_feature_mean:.4f}"r"$\pm$""{dataset_all_feature_std:.4f}", f"{dataset_all_feature_sparseness*100:.2f}"r"\%", 
                    f"{dataset_all_TFI_mean:.4f}"r"$\pm$""{dataset_all_TFI_std:.4f}", 
                    f"{dataset_G_feature_mean:.4f}"r"$\pm$""{dataset_G_feature_std:.4f}", f"{dataset_G_feature_sparseness*100:.2f}"r"\%", 
                    f"{dataset_G_TFI_mean:.4f}"r"$\pm$""{dataset_G_TFI_std:.4f}", 
                    f"{dataset_NG_feature_mean:.4f}"r"$\pm$""{dataset_NG_feature_std:.4f}", f"{dataset_NG_feature_sparseness*100:.2f}"r"\%",
                    f"{dataset_NG_TFI_mean:.4f}"r"$\pm$""{dataset_NG_TFI_std:.4f}",
                    f"{RATIO[i]*100:.0f}"r"\%"])
df = pd.DataFrame(df_data, columns=["", "Value", "Sparseness", "TFI", 
                                    "Value", "Sparseness", "TFI", 
                                    "Value", "Sparseness", "TFI", ""])
df
```

```python
latex_table = df.to_latex(index=False, 
                            escape=False, 
                            column_format="*{"+str(11)+"}{c}",
                            label="tab:rebuttal_datasets_statistics",
                            position="htbp",
                            caption="Summary Statistics of Values, TFI, and Sparseness for Selected Features.",)
lines = latex_table.splitlines()
lines.insert(2, r'\resizebox{1\hsize}{!}{')
lines.insert(6, r"\multirow{2}{*}{Dataset} & \multicolumn{3}{c}{$X$} & \multicolumn{3}{c}{$X_G$} & \multicolumn{3}{c}{$X_{\neg G}$} & \multirow{2}{*}{$r$} \\")
lines.insert(7, r"\cmidrule(lr){2-4} \cmidrule(lr){5-7} \cmidrule(lr){8-10}")
lines.insert(len(lines)-1, r'}')
latex_table_with_rules = "\n".join(lines)
print(latex_table_with_rules)
```
